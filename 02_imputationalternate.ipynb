```python
# =============================================================================
# Imports
# =============================================================================
import pathlib
import warnings
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pyreadstat
from tqdm import tqdm  # Progress bars

# Scikit-learn and related ML/Stats libraries
from sklearn.experimental import enable_iterative_imputer  # noqa: F401 # Enable experimental feature
from sklearn.impute import IterativeImputer, SimpleImputer
from sklearn.linear_model import LogisticRegression       # Binary estimator for MICE
from mord import LogisticIT                             # Ordinal estimator for MICE

warnings.filterwarnings("ignore", category=FutureWarning)
warnings.filterwarnings("ignore", category=UserWarning, module='sklearn')
```

```python
# =============================================================================
# Configuration / Constants (Imputation Stage)
# =============================================================================
# SAV_FILE_PATH = './data/WVS_Cross-National_Wave_7_inverted_sav_v6_0.sav'
# SELECTED_VARS_CSV_PATH = './data/selected_vars.csv'
# COUNTRY_COLUMN = 'B_COUNTRY_ALPHA'

# --- New Configuration for Imputation ---
DATA_DIR = pathlib.Path("./data")

# *** 'Variable_Code' and 'Type' ('binary', 'ordinal') columns. ***
VAR_TYPE_FILE_PATH = DATA_DIR / "variable_info.csv" # 

IMPUTED_OUTPUT_CSV_PATH = DATA_DIR / "wvs_wave7_imputed_alternate.csv"
DEMOGRAPHIC_GROUP_COLUMN = 'demographic_group' # Name for the generated group key column

# Note: Ensure these align with columns present after pre-processing
```

```python
# Define demographic variables used for grouping during imputation
# Note: Ensure these align with columns present after pre-processing
IMPUTATION_DEMOGRAPHIC_VARS = [
    'B_COUNTRY_ALPHA', # country
    'H_URBRURAL',      # urban / rural
    # 'Q273',          # marital status
    'Q260',            # sex
    'X003R2',          # age (3-cat)
    'Q275R',           # education (3 groups)
    # 'Q279',          # employment
```

```python
# MICE Imputation Parameters
MICE_MAX_ITER = 10
MICE_RANDOM_STATE = 42
MICE_SAMPLE_POSTERIOR = True
```

```python
# Post-Imputation Analysis Parameters
MISSING_THRESHOLDS_POST_IMPUTATION = [0.1, 1, 5, 10, 25] # Check for *any* remaining missingness
HEATMAP_CHUNK_SIZE_POST_IMPUTATION = 20
POST_IMPUTATION_SUMMARY_PLOT_PATH = DATA_DIR / 'missing_data_summary_post_imputation.png'
POST_IMPUTATION_DETAIL_PLOT_BASE = DATA_DIR / 'missing_data_heatmap_post_imputation_part_'
```

```python
# =============================================================================
# Function Definitions (Imputation Stage)
# =============================================================================

def load_variable_types(filepath: pathlib.Path) -> pd.DataFrame:
    """Loads variable type information (binary/ordinal) from a CSV file."""
    print(f"Loading variable type information from: {filepath}")
    if not filepath.exists():
        raise FileNotFoundError(
            f"Error: Variable type file not found at '{filepath}'. "
            "Please ensure the file exists and VAR_TYPE_FILE_PATH is set correctly."
        )
    var_type_df = pd.read_csv(filepath)
    # Basic validation
    if not all(col in var_type_df.columns for col in ['Variable_Code', 'Type']):
        raise ValueError(
            f"Error: Variable type file '{filepath}' must contain 'Variable_Code' and 'Type' columns."
        )
    print(f"Loaded types for {len(var_type_df)} variables.")
    var_type_df['Type'] = var_type_df['Type'].str.lower() # Standardize type names
    return var_type_df

def select_imputation_columns(df: pd.DataFrame, selected_vars_df: pd.DataFrame,
                              demographic_vars: list[str]) -> tuple[pd.DataFrame, list[str]]:
    """
    Selects the final DataFrame and identifies columns for imputation.
    Assumes `df` is the output from the pre-processing stage.
    """
    print("Selecting columns for imputation...")
    # Get survey variables marked as 'Yes' in the selection file
    survey_columns = selected_vars_df.loc[
        selected_vars_df["Selected"].str.lower() == "yes", "Variable_Code"
    ].tolist()

    # Define all columns needed: demographics + selected survey questions
    all_needed_columns = list(set(demographic_vars + survey_columns)) # Use set to avoid duplicates

    # Check if all needed columns are present in the input dataframe `df`
    missing_in_df = [col for col in all_needed_columns if col not in df.columns]
    if missing_in_df:
        print(f"Warning: The following needed columns were not found in the input DataFrame and will be excluded: {missing_in_df}")
        all_needed_columns = [col for col in all_needed_columns if col in df.columns]

    # Identify columns intended for imputation (survey columns present in the df)
    columns_to_impute = [col for col in survey_columns if col in df.columns]

    df_imputation_base = df[all_needed_columns].copy()
    print(f"Selected {df_imputation_base.shape[1]} columns for the imputation base.")
    print(f"Identified {len(columns_to_impute)} columns targeted for imputation.")

    return df_imputation_base, columns_to_impute


def prepare_mice_estimators(df: pd.DataFrame, columns_to_impute: list[str]) -> dict:
    """
    Pre-computes the appropriate scikit-learn estimator for each column
    based on its unique values (binary vs. ordinal).
    """
    print("Preparing MICE estimators based on variable types...")
    estimator_map = {}

    def make_estimator(series: pd.Series):
        """Choose a learner based on unique value set."""
        # Drop NA only for determining categories, handle cases with all NA later
        unique_values = series.dropna().unique()
        cats = sorted(unique_values)

        # Check for binary (allowing for float representations like 0.0, 1.0)
        is_binary = all(val in [0, 1, 0.0, 1.0] for val in unique_values) and len(unique_values) <= 2

        if is_binary:
            # print(f"   - Column '{series.name}': Using LogisticRegression (binary)")
            return LogisticRegression(max_iter=250, solver="lbfgs", random_state=MICE_RANDOM_STATE)
        elif len(cats) >= 2: # Treat anything with 2 or more non-binary values as ordinal for LogisticIT
             # print(f"   - Column '{series.name}': Using LogisticIT (ordinal/multi-category)")
             # Ensure LogisticIT can handle the number of categories
             # Note: LogisticIT expects integer categories. Input data should ideally be integer.
             return LogisticIT(alpha=1.0, max_iter=250) # Removed random_state if not supported or needed
        elif len(cats) == 1:
             # print(f"   - Column '{series.name}': Has only one unique value. Skipping estimator (will use default or skip in MICE).")
             return None # Let IterativeImputer handle single-valued columns (or skip them)
        else: # Only NaNs
             # print(f"   - Column '{series.name}': Contains only NaNs. Skipping estimator.")
             return None # Cannot determine type or fit estimator

    for col in columns_to_impute:
        if col in df.columns:
            estimator = make_estimator(df[col])
            if estimator: # Only add if an estimator could be determined
                 estimator_map[col] = estimator
        else:
            print(f"Warning: Column '{col}' listed for imputation not found in DataFrame during estimator prep.")

    print(f"Prepared estimators for {len(estimator_map)} columns.")
    return estimator_map


def create_demographic_groups(df: pd.DataFrame, demographic_vars: list[str], group_col_name: str) -> tuple[pd.DataFrame, list]:
    """Creates a unique key for each combination of demographic variables."""
    print(f"Creating demographic group keys using variables: {demographic_vars}")

    def create_group_key(row):
        # Handle potential NaNs in grouping variables by converting them to a string 'missing'
        vals = [
            "missing" if pd.isna(row[v]) else str(int(row[v])) if pd.api.types.is_numeric_dtype(row[v]) else str(row[v])
            for v in demographic_vars if v in row # Ensure var exists
        ]
        return "_".join(vals)

    df[group_col_name] = df.apply(create_group_key, axis=1)
    groups = df[group_col_name].unique()
    print(f"Created {len(groups)} unique demographic groups.")
    return df, groups


def post_process_imputed_ordinals(df_imputed: pd.DataFrame, df_original_group: pd.DataFrame,
                                  var_type_df: pd.DataFrame) -> pd.DataFrame:
    """Rounds and clips imputed ordinal variables to match original range."""
    ordinal_codes = var_type_df.loc[
        var_type_df["Type"] == "ordinal", "Variable_Code"
    ].tolist()
    ordinal_to_process = df_imputed.columns.intersection(ordinal_codes)

    if not ordinal_to_process.empty:
        # print(f"   - Post-processing {len(ordinal_to_process)} ordinal variables...")
        for col in ordinal_to_process:
            if col in df_original_group.columns: # Check if original column exists for range check
                 original_series = df_original_group[col].dropna()
                 if not original_series.empty:
                     min_val = original_series.min()
                     max_val = original_series.max()
                     # Round first, then clip
                     df_imputed[col] = np.round(df_imputed[col])
                     df_imputed[col] = df_imputed[col].clip(min_val, max_val).astype(original_series.dtype)
                 else: # Handle case where original column had only NaNs
                      df_imputed[col] = np.round(df_imputed[col]).astype(int) # Round and convert to int
            else: # Fallback if original column somehow missing (shouldn't happen often)
                 df_imputed[col] = np.round(df_imputed[col]).astype(int)
    return df_imputed


def impute_within_groups(df_grouped: pd.DataFrame, group_col_name: str, groups: list,
                         demographic_vars: list[str], columns_to_impute: list[str],
                         estimator_map: dict, var_type_df: pd.DataFrame) -> pd.DataFrame:
    """Performs Iterative Imputation within each demographic group."""
    print(f"\nStarting imputation process across {len(groups)} demographic groups...")
    processed_groups = []
    imputable_cols_in_scope = [col for col in columns_to_impute if col in df_grouped.columns]

    # Define the callback for IterativeImputer
    def mice_estimator_callback(col_name: str):
        """Callback to provide the pre-computed estimator for a column."""
        return estimator_map.get(col_name) # Returns None if col not in map

    # Iterate through each demographic group with a progress bar
    for group_key in tqdm(groups, desc="Processing demographic groups"):
        # print(f"\nProcessing group: {group_key}")
        group_data = df_grouped.loc[df_grouped[group_col_name] == group_key].copy()
        group_ids = group_data[demographic_vars].copy() # Keep demographic identifiers aside

        # Data for imputation (only the survey questions/columns to impute)
        X_group = group_data[imputable_cols_in_scope].copy()

        # --- Sanity Checks for the Group ---
        # 1. Skip if group is too small (e.g., <= 1 record)
        if len(group_data) <= 1:
             # print(f"   - Group {group_key}: Skipped (<= 1 record)")
             # Ensure columns match schema even if skipped
             group_data_final = pd.concat([group_ids, X_group], axis=1)
             processed_groups.append(group_data_final)
             continue

        # 2. Skip if there's nothing to impute in this group's target columns
        if not X_group.isnull().any().any():
            # print(f"   - Group {group_key}: Skipped (No missing data in target columns)")
            group_data_final = pd.concat([group_ids, X_group], axis=1)
            processed_groups.append(group_data_final)
            continue

        # --- Prepare for Imputation ---
        # 3. Identify columns entirely missing *within this group*
        all_missing_in_group = X_group.columns[X_group.isnull().all()].tolist()
        if all_missing_in_group:
            # print(f"   - Group {group_key}: Found {len(all_missing_in_group)} columns with all missing values: {all_missing_in_group}")
            X_group_imp = X_group.drop(columns=all_missing_in_group)
        else:
            X_group_imp = X_group

        # 4. Skip if no columns left after removing all-missing ones
        if X_group_imp.empty:
             # print(f"   - Group {group_key}: Skipped (No columns left after removing fully missing ones)")
             group_data_final = pd.concat([group_ids, X_group], axis=1) # Re-add original X_group with NaNs
             processed_groups.append(group_data_final)
             continue

        # --- Perform Imputation ---
        # print(f"   - Group {group_key}: Running MICE on {X_group_imp.shape[1]} variables for {X_group_imp.shape[0]} records...")
        try:
            imputer = IterativeImputer(
                estimator=None, # Will be dynamically assigned by _estimator attribute
                sample_posterior=MICE_SAMPLE_POSTERIOR,
                max_iter=MICE_MAX_ITER,
                random_state=MICE_RANDOM_STATE,
                skip_complete=True, # Skip columns with no missing values
                add_indicator=False # Do not add indicator columns
            )

            # Assign the custom estimator selection logic via the internal attribute
            # This relies on internal structure but is a known pattern for custom estimators per column.
            imputer._estimator = mice_estimator_callback

            # Ensure data is numeric for imputation
            X_group_imp_numeric = X_group_imp.apply(pd.to_numeric, errors='coerce') # Coerce non-numeric to NaN

            # Fit and transform
            imputed_data_np = imputer.fit_transform(X_group_imp_numeric)

            # Convert back to DataFrame
            X_complete_group = pd.DataFrame(
                imputed_data_np,
                columns=X_group_imp.columns,
                index=X_group_imp.index
            )

            # --- Post-Process and Combine ---
            # Round and clip ordinal variables
            X_complete_group = post_process_imputed_ordinals(X_complete_group, group_data, var_type_df)

            # Add back any columns that were all missing for this group
            if all_missing_in_group:
                for col in all_missing_in_group:
                    X_complete_group[col] = np.nan # Re-insert as NaNs

            # Re-attach demographic identifiers
            # Ensure column order is consistent if needed later, though concat handles alignment
            final_group_df = pd.concat([group_ids, X_complete_group.reindex(columns=imputable_cols_in_scope)], axis=1)
            processed_groups.append(final_group_df)
            # print(f"   - Group {group_key}: Completed processing.")

        except Exception as e:
            print(f"   - Group {group_key}: ERROR during imputation - {e}")
            print(f"     Skipping group {group_key}, original data with NaNs will be kept.")
            # Append original data for this group if imputation fails
            group_data_final = pd.concat([group_ids, X_group], axis=1)
            processed_groups.append(group_data_final)


    # --- Final Assembly ---
    if not processed_groups:
         print("Warning: No groups were processed successfully. Returning empty DataFrame.")
         return pd.DataFrame(columns=df_grouped.columns) # Return empty DF with original cols

    print("\nConcatenating results from all groups...")
    # Ensure all dataframes have the same columns before concat, adding missing ones with NaNs if needed
    all_cols = df_grouped.columns.drop(group_col_name) # Target columns
    processed_groups_aligned = []
    for df_piece in processed_groups:
         processed_groups_aligned.append(df_piece.reindex(columns=all_cols))

    imputed_data_full = pd.concat(processed_groups_aligned, ignore_index=True)
    print(f"Imputation complete! Processed {len(groups)} demographic groups.")
    print(f"Shape after group-wise imputation: {imputed_data_full.shape}")

    return imputed_data_full


def fallback_imputation(df: pd.DataFrame, demographic_vars: list[str]) -> pd.DataFrame:
    """Applies SimpleImputer (mode) to any remaining missing values."""
    print("\nApplying fallback imputation for remaining NaNs...")
    # Identify columns with missing data, excluding demographics
    cols_with_nan = df.columns[df.isnull().any()].tolist()
    cols_to_fallback_impute = [
        col for col in cols_with_nan if col not in demographic_vars
    ]

    if not cols_to_fallback_impute:
        print("No remaining NaNs found in non-demographic columns. Skipping fallback imputation.")
        return df

    print(f"Found {len(cols_to_fallback_impute)} columns for fallback mode imputation: {cols_to_fallback_impute}")

    try:
        si = SimpleImputer(strategy="most_frequent")
        df[cols_to_fallback_impute] = si.fit_transform(df[cols_to_fallback_impute])
        print("Fallback imputation completed using mode.")
    except Exception as e:
        print(f"Error during fallback imputation: {e}. Some NaNs might remain.")

    return df

def save_data(df: pd.DataFrame, filepath: pathlib.Path):
    """Saves the DataFrame to a CSV file."""
    print(f"\nSaving imputed data to: {filepath}")
    try:
        df.to_csv(filepath, index=False)
        print(f"Successfully saved {filepath} ({df.shape[0]:,} rows × {df.shape[1]} cols)")
    except Exception as e:
        print(f"Error saving data to {filepath}: {e}")

def analyze_within_country_missing(data: pd.DataFrame, country_col: str, thresholds: list[int]) -> pd.DataFrame:
    """Analyzes missing data percentages within each country for different thresholds."""
    print(f"\nAnalyzing missing data within each country for thresholds: {thresholds}%...")

    # Calculate missing data percentage by country and column
    missing_by_country = data.groupby(country_col).apply(
        lambda x: x.drop(country_col, axis=1).isnull().mean() * 100
    )

    # Identify columns meeting thresholds for each country
    country_missing_analysis = {}
    for country in missing_by_country.index:
        country_data = missing_by_country.loc[country]
        country_analysis = {f"≥{t}%_missing": country_data[country_data >= t].index.tolist() for t in thresholds}
        country_missing_analysis[country] = country_analysis

    # Create a summary table
    summary_data = []
    for country, analysis in country_missing_analysis.items():
        row = {'Country': country}
        row.update({f"≥{t}%_missing": len(analysis[f"≥{t}%_missing"]) for t in thresholds})
        summary_data.append(row)

    summary_df = pd.DataFrame(summary_data).set_index('Country')
    # Sort by the highest threshold count (Eg: 100%) descending
    summary_df = summary_df.sort_values(by=f"≥{thresholds[-1]}%_missing", ascending=False)

    print("\n=== Summary of Columns with Missing Data by Country (%) ===")
    print(summary_df)

    # Return the detailed breakdown by country/column for plotting
    return summary_df, missing_by_country

def visualize_missing_data_summary(summary_df: pd.DataFrame, filename: str = 'missing_data_summary_by_country.png'):
    """Generates a heatmap summarizing missing column counts per country."""
    print("\nGenerating summary heatmap of missing data counts...")
    plt.figure(figsize=(12, max(6, len(summary_df) * 0.3)))
    sns.heatmap(summary_df, cmap='YlOrRd', annot=True, fmt='d', linewidths=0.5)
    plt.title('Number of Columns with ≥ Threshold % Missing Data by Country')
    plt.tight_layout()
    plt.savefig(filename, dpi=300, bbox_inches='tight')
    print(f"Saved summary heatmap to: {filename}")
    plt.show()

def visualize_missing_data_details(missing_by_country_df: pd.DataFrame, chunk_size: int, base_filename: str = 'missing_data_heatmap_part_'):
    """Generates detailed heatmaps of missing percentages in chunks."""
    print(f"\nGenerating detailed heatmaps of missing percentages (chunk size: {chunk_size})...")
    num_cols = missing_by_country_df.shape[1]
    num_countries = missing_by_country_df.shape[0]

    for i in range(0, num_cols, chunk_size):
        end = min(i + chunk_size, num_cols)
        chunk_data = missing_by_country_df.iloc[:, i:end]
        part_num = i // chunk_size + 1
        filename = f"{base_filename}{part_num}.png"

        plt.figure(figsize=(min(25, chunk_data.shape[1] * 0.8), max(8, num_countries * 0.25)))
        sns.heatmap(chunk_data, cmap='YlOrRd', vmin=0, vmax=100,
                    linewidths=0.5, linecolor='lightgray', cbar_kws={'label': '% Missing Data'})
        plt.title(f'Percentage of Missing Data by Country (Columns {i+1}-{end})')
        plt.ylabel('Country Code')
        plt.xlabel('Survey Questions/Columns')
        plt.xticks(rotation=90, fontsize=8)
        plt.yticks(rotation=0, fontsize=8)
        plt.tight_layout()
        plt.savefig(filename, dpi=300, bbox_inches='tight')
        print(f"Saved detailed heatmap chunk {part_num} to: {filename}")
        plt.show()

def load_selected_variables(csv_path: str) -> pd.DataFrame:
    """Loads the list of selected variables from a CSV file."""
    print(f"Loading selected variables from: {csv_path}")
    selected_vars_df = pd.read_csv(csv_path)
    print(f"Loaded {len(selected_vars_df)} variable definitions.")
    return selected_vars_df
```

```python
# =============================================================================
# Main Execution Workflow (Imputation)
# =============================================================================

def run_imputation_pipeline(preprocessed_data_path: str,
                           selected_vars_path: str,
                           var_type_path: str,
                           output_path: str):
    """
    Orchestrates the entire imputation pipeline.

    Args:
        preprocessed_data_path: Path to the CSV/pickle file containing the data
                                after cleaning and manual imputation.
        selected_vars_path: Path to the CSV defining which variables were selected.
        var_type_path: Path to the CSV defining variable types (binary/ordinal).
        output_path: Path where the final imputed CSV will be saved.
    """
    print("--- Starting WVS Data Imputation Pipeline ---")

    # 1. Load Preprocessed Data (Assuming it was saved from previous step)
    #    Alternatively, if run in the same session, pass the DataFrame directly.
    print(f"Loading preprocessed data from: {preprocessed_data_path}")
    if not pathlib.Path(preprocessed_data_path).exists():
         raise FileNotFoundError(f"Preprocessed data file not found: {preprocessed_data_path}")
    df_preprocessed = pd.read_csv(preprocessed_data_path)
    print(f"Loaded preprocessed data shape: {df_preprocessed.shape}")

    # 2. Load Supporting Files
    selected_vars_df = load_selected_variables(selected_vars_path) # Reuse function
    var_type_df = load_variable_types(var_type_path)

    # 3. Select Columns for Imputation Base
    df_imputation_base, columns_to_impute = select_imputation_columns(
        df_preprocessed, selected_vars_df, IMPUTATION_DEMOGRAPHIC_VARS
    )

    # 4. Prepare MICE Estimators
    estimator_map = prepare_mice_estimators(df_imputation_base, columns_to_impute)

    # 5. Create Demographic Groups
    df_grouped, groups = create_demographic_groups(
        df_imputation_base, IMPUTATION_DEMOGRAPHIC_VARS, DEMOGRAPHIC_GROUP_COLUMN
    )

    # 6. Perform Imputation Within Groups
    df_imputed_groups = impute_within_groups(
        df_grouped, DEMOGRAPHIC_GROUP_COLUMN, groups,
        IMPUTATION_DEMOGRAPHIC_VARS, columns_to_impute,
        estimator_map, var_type_df
    )

    # 7. Apply Fallback Imputation (Mode)
    df_final_imputed = fallback_imputation(df_imputed_groups, IMPUTATION_DEMOGRAPHIC_VARS)

    # 8. Save Final Imputed Data
    save_data(df_final_imputed, output_path)

    # 9. Post-Imputation Missing Data Analysis (Optional but Recommended)
    print("\n--- Performing Post-Imputation Missing Data Analysis ---")
    # Check if COUNTRY_COLUMN is present before analysis
    if COUNTRY_COLUMN in df_final_imputed.columns:
        # Reuse analysis functions from the previous stage's refactoring
        # (Ensure these functions are defined or imported if running separately)
        try:
             # --- Analysis ---
             summary_df_post, details_df_post = analyze_within_country_missing(
                 df_final_imputed, COUNTRY_COLUMN, MISSING_THRESHOLDS_POST_IMPUTATION
                 ) # Use analyze_within_country_missing if defined

             # --- Visualization ---
             visualize_missing_data_summary(
                 summary_df_post,
                 filename=str(POST_IMPUTATION_SUMMARY_PLOT_PATH)
                 ) # Use visualize_missing_data_summary if defined

             visualize_missing_data_details(
                 details_df_post,
                 chunk_size=HEATMAP_CHUNK_SIZE_POST_IMPUTATION,
                 base_filename=str(POST_IMPUTATION_DETAIL_PLOT_BASE)
             ) # Use visualize_missing_data_details if defined
        except NameError as e:
             print(f"Could not run post-imputation analysis: Required analysis function not found ({e}).")
        except Exception as e:
             print(f"An error occurred during post-imputation analysis: {e}")
    else:
        print(f"Skipping post-imputation analysis by country: Column '{COUNTRY_COLUMN}' not found.")


    print("\n--- WVS Data Imputation Pipeline Finished ---")
    return df_final_imputed
```

```python
# =============================================================================
# Script Execution
# =============================================================================

if __name__ == "__main__":
    # --- Define Paths ---
    # This assumes the output of the *first* script (pre-processing) was saved.
    # If running sequentially in a notebook, you might pass the DataFrame directly.
    PREPROCESSED_DATA_FILE = DATA_DIR / "wvs_wave7_preprocessed.csv"
    SELECTED_VARS_FILE = DATA_DIR / "selected_vars.csv"
    VARIABLE_TYPES_FILE = VAR_TYPE_FILE_PATH # Defined in config block above
    IMPUTED_OUTPUT_FILE = IMPUTED_OUTPUT_CSV_PATH # Defined in config block above
    
    # Define COUNTRY_COLUMN
    COUNTRY_COLUMN = 'B_COUNTRY_ALPHA'

    # --- Execute the pipeline ---
    try:
        # Ensure the output directory exists
        IMPUTED_OUTPUT_FILE.parent.mkdir(parents=True, exist_ok=True)

        # Run the main imputation workflow
        final_imputed_df = run_imputation_pipeline(
            preprocessed_data_path=str(PREPROCESSED_DATA_FILE),
            selected_vars_path=str(SELECTED_VARS_FILE),
            var_type_path=VARIABLE_TYPES_FILE,
            output_path=IMPUTED_OUTPUT_FILE
        )
        print("\nAccess the final imputed DataFrame as `final_imputed_df` (if run in interactive session).")

    except FileNotFoundError as e:
        print(f"\nCritical Error: Input file not found. Please check paths.")
        print(e)
    except ValueError as e:
        print(f"\nCritical Error: Data validation failed.")
        print(e)
    except Exception as e:
        print(f"\nAn unexpected error occurred during the pipeline execution:")
        print(e)
        
        # For adding traceback for debugging
        # import traceback
        # traceback.print_exc()
```
